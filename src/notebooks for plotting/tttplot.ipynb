```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import genextreme, gaussian_kde, norm
```

```python
def ecdf(sample):

    # convert sample to a numpy array, if it isn't already
    sample = np.atleast_1d(sample)

    # find the unique values and their corresponding counts
    quantiles, counts = np.unique(sample, return_counts=True)

    # take the cumulative sum of the counts and divide by the sample size to
    # get the cumulative probabilities between 0 and 1
    cumprob = np.cumsum(counts).astype(np.double) / sample.size

    return quantiles, cumprob
```

```python
def tttplot(files, labels, instance, gap):
    series = []
    for file in files:
        serie = []
        f = open(file)
        for line in f:
            serie.append(float(line))
        series.append(np.array(serie))
    
    fig = plt.figure(figsize=(7,7))
    
    for idx, serie in enumerate(series):
        X = np.sort(serie)
        Y = X/X.sum()
        print(X)
        print(np.cumsum(Y))
        plt.plot(X,np.cumsum(Y), label=labels[idx], linestyle='none', marker = '+', markersize=6)
    plt.legend()
    plt.title(instance+" - Gap = "+str(gap)+"%")
    plt.xlabel("Tempo(s)")
    plt.ylabel("P[Tempo <= X]")
    plt.savefig(instance+"_"+gap+".png")
    plt.show()
        
```

```python
tttplot(["tttplot_normal_Geo_n120_ds_01_1.dat", "tttplot_reheat_Geo_n120_ds_01_1.dat"],["Sem reaquecimento","Com reaquecimento"], "Geo_n120_ds_01", "1")
```
